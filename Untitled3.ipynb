```python
def create_dly_data(X,y,dly):
    n,p=X.shape
    a,b=y.shape
    ndly=n-dly
    pdly=(dly+1)*p
    Xdly=np.zeros([ndly,pdly])
    ydly=np.zeros([ndly,b])
    for i in range(ndly):
            Xdly[i,:]=X[i:i+dly+1].reshape(-1)
            ydly[i,:]=y[i+dly,:]
    return Xdly,ydly
```

```python
Xdly,ydly=create_dly_data(X,y,dly=6)
```

```python
Xdlytr, Xdlyts, ydlytr, ydlyts = train_test_split(Xdly, ydly, test_size=0.33, random_state=0)
```

```python
regr.fit(Xdlytr,ydlytr)
ydlyhat = regr.predict(Xdlyts)
RSSdly = np.mean((ydlyhat-ydlyts)**2)/(np.std(ydlyts)**2)
Rsqdly = 1-RSSdly
print("RSS per sample = {0:f}".format(RSSdly))
print("R^2 =            {0:f}".format(Rsqdly))
```

```python
i=[0,1]
for index in range(len(i)):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,index+1)
    plt.scatter(ydlyhat[:,index],ydlyts[:,index])
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.grid()
```

```python
nred=6000
Xred=X[0:nred,:]
yred=y[0:nred,:]
```

```python

dmax = 15
Xdly,ydly=create_dly_data(Xred,yred,dmax)
```

```python
import sklearn.model_selection
import numpy.polynomial.polynomial as poly
nfold=5
kf = sklearn.model_selection.KFold(n_splits=nfold,shuffle=True)
dtest=np.arange(0,dmax)
nd=len(dtest)
Rsq=np.zeros((nd,nfold))

# Loop over the folds
for isplit, Ind in enumerate(kf.split(Xdly)):
    
    print("fold = %d " % isplit)
    
    # Get the training data in the split
    Itr, Its = Ind      
    for it, d in enumerate(dtest):
        # TODO:  
        # Xdly1 = Xdly with the columns corresponding to only the `d+1` most recent times.
        Xdly1=np.zeros([nred-d,(dmax+1)*nneuron-d])
        Xdly1=Xdly[:,d+1:]

        # TODO
        # Split the data (Xdly1,ydly) into training and test
        # Xtr = ...
        # ytr = ...
        # Xts = ...
        # yts = ...
        Xtr = Xdly1[Itr]
        ytr = ydly[Itr]
        Xts = Xdly1[Its]
        yts = ydly[Its]
        regr.fit(Xtr,ytr,d)
        ytshat = regr.predict(Xts)
        Rsq[it,isplit] = -np.mean((ytshat-yts)**2)/(np.std(ydlyts)**2)+1
        # TODO:  Fit data on training data
        
        # TODO:  Measure the R^2 vale on test data and store in the matrix Rsq
        
print(Rsq)
```

```python
RSS_mean = np.mean(Rsq,axis=1) #note mean is taken over the second dimension, i.e. all folds for the same model order
RSS_std  = np.std(Rsq,axis=1)/np.sqrt(nfold-1)
plt.errorbar(dtest, RSS_mean, yerr=RSS_std, fmt='-')
plt.ylim(0,0.8)
plt.xlabel('Model order')
plt.ylabel('Test RSS')
plt.grid()
```

```python
imax = np.argmax(RSS_mean)
print("The selected model order is {0:d}".format(dtest[imax]))
```

```python

RSS_tgt = RSS_mean[imax] - RSS_std[imax]

# Find the lowest model order below the target
I = np.where(RSS_mean >= RSS_tgt)[0]
iopt = I[0]
dopt = dtest[iopt]
print('optimal model order with one SE rule= %d' % dopt)
print(RSS_mean[iopt])
```
